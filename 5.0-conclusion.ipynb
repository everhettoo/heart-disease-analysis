# Objective
To display the result of all the model.

```python
import matplotlib.pyplot as plt
import custom_libs.evaluation_helper as ev

# The dummy values was used for testing.
# ev.accuracies = { 'nb':80, 'dt':70, 'lr': 90, 'svm':88, 'rf':90 }

#  Sort the values from lowest to highest.
ev.accuracies  = dict(sorted(ev.accuracies.items(), key=lambda item: item[1]))

# Function to display value on the bar chart.
def add_labels(x_pos, y_pos):
    for i in range(len(x_pos)):
        plt.text(i, y_pos[i], y_pos[i], ha = 'center')

bar_width = 0.35

# To print values for debugging.
print(f'accuracies          = {ev.accuracies}')
print(f'accuracies (key)    = {list(ev.accuracies.keys())}')
print(f'accuracies (values) = {list(ev.accuracies.values())}')

# Draw the bar chat and display the values.
plt.figure(figsize=(7,7))
plt.title('Model Accuracy Comparison')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.bar(range(len(ev.accuracies)), list(ev.accuracies.values()), align='center', width=bar_width)
add_labels(list(ev.accuracies.keys()), list(ev.accuracies.values()))
plt.xticks(range(len(ev.accuracies)), list(ev.accuracies.keys()))
plt.show()
```
